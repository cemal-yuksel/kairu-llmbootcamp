# 8. Pratik: Python ile LLM Kullanımı Adım Adım

Bu bölümde, Python kullanarak bir LLM (ör. OpenAI GPT) ile nasıl uygulama geliştireceğimizi adım adım öğreneceğiz.

## 8.1. Gerekli Kütüphanelerin Kurulumu

LLM API'lerini kullanmak için genellikle `openai` gibi bir kütüphaneye ihtiyacımız olur. Kurulum için:

```bash
pip install openai
```

## 8.2. API Anahtarının Alınması ve Kullanılması

Birçok LLM servisi (ör. OpenAI), kimlik doğrulama için API anahtarı gerektirir. Bu anahtarı ilgili platformdan alıp gizli tutmalısınız.

```python
import openai

# API anahtarınızı güvenli şekilde saklayın
openai.api_key = "YOUR_API_KEY"
```

## 8.3. Basit Bir LLM Çağrısı

Aşağıda, bir metni özetleyen basit bir LLM çağrısı örneği verilmiştir:

```python
response = openai.Completion.create(
    engine="text-davinci-003",
    prompt="Bir metni 3 cümleyle özetle: Yapay zeka, insan benzeri görevleri yerine getirebilen bilgisayar sistemleridir. Son yıllarda büyük ilerlemeler kaydedilmiştir. Bu teknolojiler, sağlık, eğitim ve finans gibi birçok alanda kullanılmaktadır.",
    max_tokens=100,
    temperature=0.7
)
print(response.choices[0].text.strip())
```

### Açıklama:
- `engine`: Kullanılan modelin adı.
- `prompt`: Modele verdiğimiz giriş metni.
- `max_tokens`: Yanıtın uzunluğunu belirler.
- `temperature`: Yanıtın çeşitliliğini kontrol eder (0: deterministik, 1: daha yaratıcı).

## 8.4. Fonksiyon Halinde LLM Çağrısı

LLM çağrılarını tekrar tekrar kullanmak için fonksiyon haline getirebiliriz:

```python
def metni_ozetle(metin):
    prompt = f"Bir metni 3 cümleyle özetle: {metin}"
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=100,
        temperature=0.7
    )
    return response.choices[0].text.strip()

# Kullanım
metin = "Yapay zeka, insan benzeri görevleri yerine getirebilen bilgisayar sistemleridir. Son yıllarda büyük ilerlemeler kaydedilmiştir. Bu teknolojiler, sağlık, eğitim ve finans gibi birçok alanda kullanılmaktadır."
ozet = metni_ozetle(metin)
print(ozet)
```

## 8.5. Prompt Tasarımında Dikkat Edilecekler

- **Açık ve net olun:** Modelden ne istediğinizi doğrudan belirtin.
- **Örnek verin:** Modelin daha iyi anlaması için örnekler ekleyin.
- **Kısa ve öz tutun:** Gereksiz detaylardan kaçının.

## 8.6. Hata Yönetimi ve Güvenlik

API çağrılarında hata olasılığına karşı kodunuzu try-except bloklarıyla güvenli hale getirin:

```python
try:
    ozet = metni_ozetle(metin)
    print(ozet)
except Exception as e:
    print(f"Bir hata oluştu: {e}")
```

## 8.7. Token ve Maliyet Yönetimi

- **Token:** Modelin yanıt üretirken kullandığı birimdir. Uzun promptlar ve yanıtlar daha fazla token harcar.
- **Maliyet:** API servislerinde token başına ücretlendirme olabilir. Kullanımınızı takip edin.

## 8.8. Uygulama Örneği: Soru-Cevap Sistemi

Aşağıda, kullanıcıdan bir soru alıp LLM ile yanıtlayan basit bir uygulama örneği verilmiştir:

```python
def soru_cevapla(soru):
    prompt = f"Aşağıdaki soruya kısa ve net bir yanıt ver: {soru}"
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=60,
        temperature=0.5
    )
    return response.choices[0].text.strip()

soru = input("Sorunuzu yazın: ")
cevap = soru_cevapla(soru)
print("Yanıt:", cevap)
```

<h3 style="color:#0288d1;">🔵 9.1. Zero-Shot Prompting <span style='font-size:0.8em;'>(01_zero_shot.py)</span></h3>
<b>Tanım:</b> <span style="color:#0288d1;"><b>Zero-shot prompting</b></span>, modele <b>hiçbir örnek vermeden</b> yalnızca görev tanımı ile komut vermektir. Model, verilen görevi daha önce gördüğü örneklerden ve genel bilgisinden yola çıkarak yerine getirir. <b>Kullanım Alanları:</b>
<ul>
<li>🔹 <b>Duygu Analizi:</b> Bir metnin olumlu, olumsuz veya nötr olup olmadığını belirleme.</li>
<li>🔹 <b>Çeviri:</b> Metni bir dilden başka bir dile çevirme.</li>
<li>🔹 <b>Özetleme:</b> Uzun bir metni kısa ve öz şekilde özetleme.</li>
<li>🔹 <b>Sınıflandırma:</b> Metni kategoriye ayırma (ör. haber türü, konu başlığı).</li>
<li>🔹 <b>Bilgi Çıkarımı:</b> Metinden anahtar bilgi veya isimleri çekme.</li>
<li>🔹 <b>Soru-Cevap:</b> Bir metne dayalı olarak soruları yanıtlama.</li>
</ul>
<b>Avantajları:</b>
<ul>
<li>⚡ <b>Hızlı ve Kolay:</b> Ekstra örnek hazırlamadan hemen kullanılabilir.</li>
<li>🌍 <b>Genel Amaçlı:</b> Çok farklı görevlerde denenebilir.</li>
</ul>
<b>Sınırlamaları:</b>
<ul>
<li>❗ <b>Bağlam Eksikliği:</b> Model, görevi yanlış anlayabilir veya beklenmedik sonuçlar üretebilir.</li>
<li>❗ <b>Doğruluk Düşebilir:</b> Özellikle karmaşık veya çok spesifik görevlerde örnek verilmediği için hata oranı artabilir.</li>

<b>Tanım:</b> <span style="color:#512da8;"><b>Chain of Thought (CoT) prompting</b></span>, modelin <b>adım adım düşünmesini</b> ve çözüm üretmesini sağlayan bir tekniktir. Özellikle karmaşık, çok adımlı veya mantıksal akıl yürütme gerektiren görevlerde kullanılır. Modelden sadece cevabı değil, <b>ara adımları</b> da açıklamasını isteriz.<br><br>
<b>Kullanım Alanları:</b>
<ul>
<li>🔹 <b>Matematik Problemleri:</b> Adım adım çözüm gerektiren hesaplamalar.</li>

<li>🔹 <b>Metin Analizi:</b> Bir metnin farklı yönlerini sistematik olarak değerlendirme.</li>
<li>🔹 <b>Problem Çözme:</b> Birden fazla sorunu analiz edip çözüm önerileri geliştirme.</li>
<li>🔹 <b>Karar Verme:</b> Alternatifleri karşılaştırıp en iyi seçeneği seçme.</li>
</ul>
<b>Avantajları:</b>
<ul>
<li>🧩 <b>Şeffaflık:</b> Modelin nasıl düşündüğünü ve sonuca nasıl ulaştığını görebilirsiniz.</li>
<li>🔍 <b>Hata Ayıklama:</b> Yanlış adımda nerede hata yapıldığını kolayca tespit edebilirsiniz.</li>
</ul>
<b>Sınırlamaları:</b>
<ul>
<li>⏳ <b>Uzun Yanıtlar:</b> Adım adım açıklama, yanıtın uzunluğunu artırabilir.</li>
<li>❗ <b>Her Görev İçin Uygun Değil:</b> Basit görevlerde gereksiz yere karmaşıklık katabilir.</li>
</ul>
<div style="background:#fffde7; border-radius:8px; padding:10px; margin:10px 0;">
<b>Uygulama Örnekleri:</b>
<ol>
<li><b>Matematik Problemi:</b><br><code>Bu matematik problemini adım adım çöz: Problem: Bir mağazada 25 TL'lik bir ürün %20 indirim yapılıyor. Daha sonra indirimi ürün %15 KDV ekleniyor. Son fiyat ne kadar olur? Adım adım çözüm:</code></li>
<li><b>Mantıksal Akıl Yürütme:</b><br><code>Bu mantık problemini adım adım analiz et: Problem: Ahmet, Mehmet'ten daha uzun... En uzundan en kısaya doğru sıralama nasıl olur? Adım adım analiz:</code></li>

<li><b>Karar Verme:</b><br><code>Bu karar verme sürecini adım adım analiz et: Durum: Yazılım geliştiricisi olarak yeni bir projede teknoloji seçimi yapman gerekiyor... Karar verme süreci:</code></li>
</ol>
</div>
<b>İpucu:</b> <span style="color:#512da8;">Chain of Thought</span> ile modelin düşünce zincirini görebilir, karmaşık problemlerde daha güvenilir sonuçlar elde edebilirsiniz.

<h3 style="color:#388e3c;">🟢 9.4. Role-Based Prompting <span style='font-size:0.8em;'>(04_role_based.py)</span></h3>
<b>Tanım:</b> <span style="color:#388e3c;"><b>Role-based prompting</b></span>, modele <b>belirli bir rol veya uzmanlık</b> vererek, o rolün bakış açısıyla yanıt vermesini sağlamaktır. Model, bir öğretmen, pazarlama uzmanı, mali müşavir, psikolog veya şef gibi davranabilir.<br><br>
<b>Kullanım Alanları:</b>
<ul>
<li>🔹 <b>Pazarlama Uzmanı:</b> Marka stratejisi, sosyal medya önerileri.</li>
<li>🔹 <b>Teknik Yazar:</b> API dokümantasyonu, kullanıcı kılavuzu yazımı.</li>
<li>🔹 <b>Mali Müşavir:</b> Finansal analiz, yatırım tavsiyesi.</li>
<li>🔹 <b>Öğretmen:</b> Eğitim materyali, konu anlatımı, örnek soru hazırlama.</li>
<li>🔹 <b>Psikolog:</b> Stres yönetimi, kişisel gelişim tavsiyeleri.</li>
<li>🔹 <b>Şef:</b> Yemek tarifi, menü önerisi.</li>
</ul>
<b>Avantajları:</b>
<ul>
<li>👔 <b>Uzmanlık ve Tutarlılık:</b> Yanıtlar daha profesyonel ve bağlama uygun olur.</li>
<li>🎭 <b>Kişiselleştirme:</b> Kullanıcıya özel, rol tabanlı deneyim sunar.</li>
</ul>
<b>Sınırlamaları:</b>
<ul>
<li>❗ <b>Rol Tanımı Net Olmalı:</b> Belirsiz veya çelişkili rol tanımları modelin kafasını karıştırabilir.</li>
</ul>
<div style="background:#fff3e0; border-radius:8px; padding:10px; margin:10px 0;">
<b>Uygulama Örnekleri:</b>
<ol>
<li><b>Pazarlama Uzmanı:</b><br><code>System: Sen deneyimli bir dijital pazarlama uzmanısın...<br>User: Yeni açılan kahve dükkanım için sosyal medya stratejisi önerir misin?</code></li>
<li><b>Teknik Yazar:</b><br><code>System: Sen profesyonel bir teknik yazarsın...<br>User: REST API'nin temel kavramlarını açıklar mısın?</code></li>
<li><b>Mali Müşavir:</b><br><code>System: Sen sertifikalı bir mali müşavirsin...<br>User: Startup'ım için yatırımcı bulmak istiyorum. Mali tablolarımda hangi metriklere odaklanmalıyım?</code></li>
<li><b>Öğretmen:</b><br><code>System: Sen deneyimli bir matematik öğretmenisin...<br>User: Lise öğrencilerine logaritma konusunu nasıl açıklayabilirim?</code></li>
<li><b>Psikolog:</b><br><code>System: Sen lisanslı bir klinik psikologsun...<br>User: İş hayatında stresle başa çıkma konusunda zorlanıyorum. Hangi teknikleri uygulayabilirim?</code></li>
<li><b>Şef:</b><br><code>System: Sen Michelin yıldızlı restoran deneyimi olan profesyonel bir şefsin...<br>User: Evde kolayca yapabileceğim, misafirlere ikram edebileceğim şık bir tatlı tarifi önerir misin?</code></li>
</ol>
</div>
<b>İpucu:</b> <span style="color:#388e3c;">Role-based prompting</span> ile modelin uzmanlık seviyesini ve yanıt kalitesini artırabilirsiniz.
</div>

<h3 style="color:#fbc02d;">🟡 9.5. OpenAI ChatCompletion API Kullanımı <span style='font-size:0.8em;'>(05_chatcompletion_api.py)</span></h3>
<b>Tanım:</b> <span style="color:#fbc02d;"><b>OpenAI ChatCompletion API</b></span>, LLM ile sohbet tabanlı uygulamalar geliştirmenizi sağlayan güçlü bir arayüzdür. Farklı parametrelerle modelin davranışını, yanıt çeşitliliğini ve çıktının kontrolünü sağlayabilirsiniz.<br><br>
<b>Kullanım Alanları:</b>
<ul>
<li>🔹 <b>System Mesajı:</b> Modelin genel davranışını ve rolünü belirleme.</li>
<li>🔹 <b>Konuşma Geçmişi:</b> Çok adımlı diyaloglarda bağlamı koruma.</li>
<li>🔹 <b>Temperature:</b> Yanıtların deterministik (düşük) veya yaratıcı (yüksek) olmasını ayarlama.</li>
<li>🔹 <b>Streaming:</b> Yanıtı parça parça (chunk) olarak alma.</li>
<li>🔹 <b>Birden Fazla Yanıt:</b> Aynı prompt için birden fazla alternatif yanıt üretme.</li>
<li>🔹 <b>Token Kullanımı:</b> Yanıt uzunluğunu ve maliyeti kontrol etme.</li>
<li>🔹 <b>Model Karşılaştırma:</b> Farklı modellerin çıktısını test etme.</li>
</ul>
<b>Avantajları:</b>
<ul>
<li>⚙️ <b>Esneklik:</b> Farklı parametrelerle modelin davranışını ince ayar yapabilirsiniz.</li>
<li>📈 <b>Performans Takibi:</b> Token kullanımı ve maliyet analizi yapabilirsiniz.</li>
</ul>
<b>Sınırlamaları:</b>
<ul>
<li>💰 <b>Maliyet:</b> Uzun yanıtlar ve yüksek token kullanımı maliyeti artırabilir.</li>
</ul>
<div style="background:#e3f2fd; border-radius:8px; padding:10px; margin:10px 0;">
<b>Uygulama Örnekleri:</b>
<ol>
<li><b>System Mesajı ile Rol Belirleme:</b><br><code>messages=[{"role": "system", "content": "Sen bir Python öğretmeni olan yardımcısın..."}, ...]</code></li>
<li><b>Konuşma Geçmişi Yönetimi:</b><br><code>conversation = [{"role": "system", ...}, {"role": "user", ...}, ...]</code></li>
<li><b>Temperature ile Yaratıcılık:</b><br><code>temperature=0.1 (daha tutarlı), temperature=1.2 (daha yaratıcı)</code></li>
<li><b>Streaming Yanıt:</b><br><code>stream=True ile yanıtı parça parça alabilirsiniz.</code></li>
<li><b>Birden Fazla Yanıt:</b><br><code>n=3 ile aynı prompt için 3 farklı yanıt alın.</code></li>
<li><b>Token Kullanımı Takibi:</b><br><code>response.usage.prompt_tokens, response.usage.completion_tokens</code></li>
<li><b>Model Karşılaştırma:</b><br><code>model="gpt-3.5-turbo" ve model="gpt-4" ile aynı promptu test edin.</code></li>
</ol>
</div>
<b>İpucu:</b> <span style="color:#fbc02d;">ChatCompletion API</span> ile uygulamanızın çıktısını ve maliyetini optimize edebilirsiniz.
</div>
